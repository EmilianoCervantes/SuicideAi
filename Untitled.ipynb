```python
corpus_raw = u""
for archivo in archivos:
    print("Estoy leyendo el archivo {0}".format(archivo))
    with codecs.open(archivo, "r", "utf-8") as archivo:
        corpus_raw += archivo.read()
    print("El corpus tiene {0} characteres".format(len(corpus_raw)))
    print()
```

```python
tokenizer = nltk.data.load('tokenizers/punkt/spanish.pickle')
```

```python
raw_sentences = tokenizer.tokenize(corpus_raw)
```

```python
# Limpia characteres, mantiene preguntas, acentos y ñ para terminos de contexto
toktok = ToktokTokenizer()
preposiciones = ["a", "ante", "bajo", "cabe", "con", "contra", "de", "desde", "durante", "en", "entre", "hacia", "hasta", "mediante", "para", "por", "segun", "sin", "so", "sobre", "tras", "versus","via"]
articulos = ["el", "la", "los", "las", "esas","esa", "nos", "tan", "estas","ella","misma" ,"ello", "algo","cosa","pero", "como", "esta", "eres", "esas"]
stop = set(stopwords.words('spanish'))
def limpia_impureza(raw):
    texto = re.sub("[^a-zA-Zñáéíóúü]"," ",raw)
    texto = texto.lower()
    unaccented_string = unidecode.unidecode(texto).split()
    result = list(set(unaccented_string) - set(preposiciones)- set(articulos))
    final = []
    for word in result:
        if len(word) > 3:
            final.append(word)
    return toktok.tokenize(" ".join(final))
```

```python
sentences = []
toktok = ToktokTokenizer()
for raw_sentence in raw_sentences:
    if len(raw_sentence) > 0:
        sentences.append(limpia_impureza(raw_sentence))
```

```python
sentences
```

```python
token_count = sum([len(sentence) for sentence in sentences])
print("Los textos contienen {0:,} tokens".format(token_count))
```

```python
#Numero de dimensiones que va a ejecutar
#mas = a mas generalizado ya que es mas complejo
num_features =300

#Número minimo de palabras a tomar en cuenta
min_word_count=3

#Numero de threads a correr en parallelo
workers = multiprocessing.cpu_count()

#Tamaño de tu ventana cuantas palabras a la vez
context_size = 7

#Agrupamiento de palabras frecuentes a descartar
downsampling = 1e-3

#Posición aleatorio para empezar a vectorizar
#Determinista
seed = 1

#Numero de interaciones a considerar
iter=3 
#Implementa la funcion de softmax
hs=1
negative=0 
```

```python
entrenamiento = w2v.Word2Vec(sentences,
    sg=1,
    seed=seed,
    workers=workers,
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample=downsampling,
    iter=3,
    hs=1,
    negative=0
)
```

```python
def plot_region(x_bounds, y_bounds):
    slice = points[
        (x_bounds[0] <= points.x) &
        (points.x <= x_bounds[1]) & 
        (y_bounds[0] <= points.y) &
        (points.y <= y_bounds[1])
    ]
        
    ax = slice.plot.scatter("x", "y", s=35, figsize=(10, 8))
    for i, point in slice.iterrows():
        ax.text(point.x + 0.005, point.y + 0.005, point.word, fontsize=11)
```

```python
plot_region(x_bounds=(-2,5, 0.0), y_bounds=(0.0, 2.0))
```
